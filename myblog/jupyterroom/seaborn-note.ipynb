# Seaborn绘图类型了解

以绘图形式，可将seaborn的绘图API分为以下几类：
* Relational plots：关系绘图
    * relplot()：关系绘图汇总接口
    * scatterplot()：散点图
    * lineplot()：折线图
* Categorical plots：分类绘图
    * catplot()：分类绘图汇总接口
    * stripplot()：分类散点图
    * swarmplot()：分类散点图
    * boxplot()：箱型图
    * violinplot()：箱型图（小提琴图）
    * boxenplot()：箱型图
    * pointplot()：点图
    * barplot()：条形图
    * countplot()：计数图
* Distribution plots：分布绘图
    * jointplot()：双变量图
    * pairplot()：多变量图
    * distplot()：分布图
    * kdeplot()：核密度估计图
    * rugplot()
* Regression plots：回归绘图
    * lmplot()
    * regplot()
    * residplot()
* Matrix plots：矩阵绘图
    * heatmap()：热力图
    * clustermap()

因为不同的绘图是使用在不同属性的变量之上的，所以先简单回顾下**数据类型的度量**
* 定性
    * 定类 —— 如 性别：男性、女性 (可用**独热编码**，但序号编码不建议)
    * 定序 —— 如 文化程度：大学、高中、初中、小学、文盲（可用独热编码，也可用**序号编码**）
* 定值
    * 定距 —— **没有一个真正的零点**。如：摄氏温度，摄氏40度比30度高10度，摄氏30度比20度又高10度，它们之间高出的距离相等，而摄氏零度并不是没有温度。
    * 定比 —— **具有一个真正的零点，因而它具有乘与除（×、÷）的数学特质**。如：年龄和收入，是定距变量，同时又是定比变量，因为其零点是绝对的，可以作乘除的运算。**定比变量一版同时也是定距变量**。
* 日期/时间

# 不常用绘图简介

* stripplot()：散点图，分类变量使用的散点图
* swarmplot()：散点图，调整的swarmplot，没有重叠
* violinplot()：小提琴图，由箱型图变异而来
* boxenplot()：箱型图，美化过的
* pointplot()：
    * 点图，代表散点图位置的数值变量的中心趋势估计，并使用误差线提供关于该估计的不确定性的一些指示。点图可能比条形图更有用于聚焦一个或多个分类变量的不同级别之间的比较。他们尤其善于表现交互作用：一个分类变量的层次之间的关系如何在第二个分类变量的层次之间变化。连接来自相同色调等级的每个点的线允许交互作用通过斜率的差异进行判断，这比对几组点或条的高度比较容易。
* rugplot()：辅助kdeplot()的方法绘图

## countplot

```python
fig = plt.figure(figsize=(18,4))
axes= fig.subplots(1,3)
sns.countplot(x="Sex",data=data_raw,
              color='r',ax=axes[0])
sns.countplot(x="Sex",hue="Survived",data=data_raw,
              color='r',ax=axes[1])
sns.countplot(y="Sex",hue="Survived",data=data_raw,
              color='r',ax=axes[2])
```

## barplot

```python
fig = plt.figure(figsize=(12,4))
axes= fig.subplots(1,2)
sns.barplot(x="Sex",y='Fare',hue="Survived",data=data_raw, # 默认是均值
            color='r',ax=axes[0]) 
from numpy import sum, median, mean, var, std
sns.barplot(x="Sex",y='Fare',hue="Survived",data=data_raw, estimator=median,
            color='r',ax=axes[1])
```

## boxplot

```python
fig = plt.figure(figsize=(20,5))
axes= fig.subplots(1,4)
sns.boxplot(x=data_raw['Age'],color='r',ax=axes[0])
sns.boxplot(y=data_raw['Age'],color='r',ax=axes[1])
sns.boxplot(x="Sex",y='Fare',hue="Survived",data=data_raw,color='r',ax=axes[2])
sns.boxplot(y="Sex",x='Fare',hue="Survived",data=data_raw,color='r',ax=axes[3])
```

## stripplot

```python
fig = plt.figure(figsize=(18,5))
axes= fig.subplots(1,3)
sns.boxplot(x="Sex",y='Fare',hue="Survived",data=data_raw,color='r',ax=axes[0])
sns.stripplot(x="Sex", y="Fare",hue="Survived", data=data_raw,color='r',ax=axes[1])
sns.swarmplot(x="Sex", y="Fare",hue="Survived", data=data_raw,color='r',ax=axes[2])
```

## swarmplot

```python
fig = plt.figure(figsize=(20,5))
axes= fig.subplots(1,4)
sns.stripplot(x="Sex", y="Fare",hue="Survived", data=data_raw,color='r',ax=axes[0],dodge=False)
sns.stripplot(x="Sex", y="Fare",hue="Survived", data=data_raw,color='r',ax=axes[1],dodge=True)
sns.swarmplot(x="Sex", y="Fare",hue="Survived", data=data_raw,color='r',ax=axes[2],dodge=False)
sns.swarmplot(x="Sex", y="Fare",hue="Survived", data=data_raw,color='r',ax=axes[3],dodge=True)
```

## distplot

```python
fig = plt.figure(figsize=(20,8))
axes= fig.subplots(2,4)
sns.distplot(data_raw['Age'].dropna(), rug=False, hist=False,
             color='r',ax=axes[0,0])
sns.distplot(data_raw['Age'].dropna(), rug=True, hist=False,
             color='r',ax=axes[0,1])
sns.distplot(data_raw['Age'].dropna(), rug=False, hist=True,
             color='r',ax=axes[0,2])
sns.distplot(data_raw['Age'].dropna(), rug=True, hist=True,
             color='r',ax=axes[0,3])
sns.distplot(data_raw['Age'].dropna(), rug=False, hist=False,
             vertical=True, color='r',ax=axes[1,0])
sns.distplot(data_raw['Age'].dropna(), rug=True, hist=False,
             vertical=True, color='r',ax=axes[1,1])
sns.distplot(data_raw['Age'].dropna(), rug=False, hist=True,
             vertical=True, color='r',ax=axes[1,2])
sns.distplot(data_raw['Age'].dropna(), rug=True, hist=True,
             vertical=True, color='r',ax=axes[1,3])
```
